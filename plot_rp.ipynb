
# Recurrence Plot


A recurrence plot is an image obtained from a time series, representing the
distances between each time point. The image can be binarized using a
threshold. It is implemented as :class:`pyts.image.RecurrencePlot`.


```python
# Author: Johann Faouzi <johann.faouzi@gmail.com>
# License: BSD-3-Clause

import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot
from pyts.datasets import load_gunpoint
X, _, _, _ = load_gunpoint(return_X_y=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

def MinMax(data):
    MMS = MinMaxScaler().fit(data)
    scaled = MMS.transform(data)
    return scaled

# print(X[0].shape)
# scaled_data = MinMax(X[0])
# print(scaled_data.shape)
# scaled_data = []
```

```python
print(np.shape(X))
X = [list(x) for x in zip(*X)]
print(np.shape(X))
X = MinMax(X)
print(np.min(X),np.max(X))
```

```python


# Recurrence plot transformation
rp = RecurrencePlot(percentage=30)
# threshold='point'
X_rp = rp.fit_transform(X)

# Show the results for the first time series
plt.figure(figsize=(5, 5))
plt.imshow(X_rp[20],cmap='binary', origin='lower')
# cmap='binary',
plt.title('Recurrence Plot', fontsize=16)
plt.tight_layout()
plt.show()
```
