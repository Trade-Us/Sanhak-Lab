```python
# Load Dataset
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons


# Needed Library!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 삽입!
import Cluster as c
```

```python
x, y = make_moons(n_samples=200, noise=0.05, random_state=0)
```

```python
df = pd.DataFrame(x)
print(df.head(10)) # 데이터 형식

# x: 2차원 특징 배열, 
# 클러스터 개수
result, scaled_x = c.kmeans(x, 2)
# x 값 , 예측 레이블 리스트, 클러스터 넘버
c.visualization_clusters(scaled_x, result.labels_, [0, 1])
```

```python
# 클래스 번호 리스트 , 클래스당 배치 사이즈
y = c.getClassLabelFor([0, 1], 100)
# 정답 레이블, 예측 레이블 (클래스 분류 1차원 배열)
accuracy = c.getAccuracy(y, result.labels_)
```
