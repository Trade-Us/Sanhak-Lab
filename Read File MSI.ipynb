```python
# columns 와 value는 사용자 입력
# df = pd.read_csv('resources/AXISX_resample.csv')
# columns = ['chip', 'wire']
# value = ['value']
df = pd.read_csv('resources/Dataset1.csv')
columns = ['Process', 'Step']
value = ['Value']
df.head()
```

```python
df = df.loc[:, columns + value] #('chip', 'wire', 'value')는 사용자 입력
df.head()
```

```python
params_list = []
for param in columns:
    df_params = df.loc[:, [param]].drop_duplicates().reset_index().drop(columns='index').transpose()
    params_list.append(df_params.values[0].tolist())
    
#df_params.head(100)
print(params_list)
```

```python
import pandas as pd
import numpy as np
from itertools import product
from itertools import chain
import copy

parameter_combination_sets = list(product(*params_list))
print(parameter_combination_sets)
```

```python
def split_into_values(df, columns, params_comb_sets):
    result_df = pd.DataFrame()

    for param_set in params_comb_sets:
        part_df = df.copy()
        for i, param in enumerate(param_set):
            part_df = part_df.loc[part_df[columns[i]] == param, :]
        part_df = part_df.reset_index().drop(columns=['index']+columns).transpose()
        result_df = result_df.append(part_df, ignore_index=True)
    return result_df[pd.notnull(result_df[0])]
```

```python
result = split_into_values(df, columns, parameter_combination_sets)
```
